```python
def load_text_files(file, path):    
    with open(os.path.join(path, file), 'r', encoding='utf-8', errors='ignore') as f:
        text = f.read()

    # text = text.lower()

    return text
```

```python
def tokenize_text(text):
    tokens = nltk.word_tokenize(text)

    # Filter out punctuation and convert to lowercase in one pass
    tokens = [token.lower() for token in tokens]
    return tokens
```

```python
def check_categories(text, unique_categories):
    """Extract wiki [[Category:...]] values from raw text and append to unique_categories.

    Uses a regex that handles single or double brackets and is case-insitive.
    Returns the updated unique_categories list.
    """

    # Match [[Category:...]] (captures the inner content up to the first closing ]])
    pattern = re.compile(r"\[\[\s*Category\s*:\s*(.+?)\s*\]\]", flags=re.IGNORECASE)
    matches = pattern.findall(text)

    for m in matches:
        cat = m.strip()
        # normalize whitespace inside category
        cat = ' '.join(cat.split())
        if cat not in unique_categories:
            unique_categories.append(cat)

    return unique_categories
```

```python
page_names = []
page_texts = {}
page_tokens = {}
unique_categories = []

for file in pages:
    page_name = file.replace(".txt", "")
    page_names.append(page_name)

    text = load_text_files(file,path)
    page_texts[page_name] = text

    tokens = tokenize_text(text)
    page_tokens[page_name] = tokens

    # Use the raw text for a robust category extraction
    unique_categories = check_categories(text, unique_categories)
```

```python
# Normalize and map categories into unified forms (handles '|' and cast variants)
import os, json
from collections import defaultdict

# canonical show names (lowercase keys used for matching)
shows = {
    'game of thrones': 'Game of Thrones',
    'house of the dragon': 'House of the Dragon',
    'a knight of the seven kingdoms': 'A Knight of the Seven Kingdoms',
}

# helper: detect if a string contains a season expression like 'Season 2' or 'season 10'
season_re = re.compile(r'season\s*(\d+)', flags=re.IGNORECASE)


def normalize_category(cat):
    """Normalize a raw category string to a canonical form.

    Rules implemented:
    - If '|' present, drop anything after '|' unless the left side contains a 'Season N' AND the right side is a numeric episode: then preserve episode as "Episode N".
    - Handle 'Seasons of <Show>|N'  '<Show>  Season N'
    - Normalize whitespace and remove surrounding template/bracket chars.
    - Detect cast-related categories via word-boundary matches for 'cast' or 'starring' (avoids matching 'podcast' or 'castle').
    - If category is cast-related and one of the known shows is mentioned, map to '<Show>  Cast' or '<Show>  Season N  Cast' (preserve season if present).
    - Otherwise return a cleaned base category (no pipe suffix).
    """
    if not cat:
        return cat

    parts = cat.split('|', 1)
    left_raw = parts[0].strip()
    after = parts[1].strip() if len(parts) > 1 else None

    base = left_raw.strip(" []{}")
    base = ' '.join(base.split())
    lower = base.lower()

    # detect season (singular)
    season_m = season_re.search(lower)
    season_part = None
    if season_m:
        season_part = f"Season {season_m.group(1)}"

    # detect whether this is a cast-like category (word boundary to avoid 'podcast' etc.)
    is_cast = bool(re.search(r'\bcast\b', lower) or re.search(r'\bstarring\b', lower))

    # find which known show (if any) is referenced
    show_found = None
    for key, canonical in shows.items():
        if key in lower:
            show_found = canonical
            break

    # Handle 'Seasons of <Show>|N'  '<Show>  Season N'
    if re.search(r'\bseasons\s+of\b', lower, re.IGNORECASE) and after and re.fullmatch(r'\d+', after):
        season_num = int(after)
        if show_found:
            return f"{show_found}  Season {season_num}"
        else:
            # Generic case: "Seasons of X|2"  "Seasons of X  Season 2"
            return f"{base}  Season {season_num}"

    # Preserve numeric episode after '|' only when season is present on the left side
    if season_part and after and re.fullmatch(r'\d+', after):
        ep = int(after)
        if show_found:
            return f"{show_found}  {season_part}  Episode {ep}"
        else:
            # attach Episode N to the base left side if no known show matched
            return f"{base}  Episode {ep}"

    # detect cast with known show
    if is_cast and show_found:
        if season_part:
            return f"{show_found}  {season_part}  Cast"
        return f"{show_found}  Cast"

    # handle 'starring' fallback
    if 'starring' in lower and show_found:
        if season_part:
            return f"{show_found}  {season_part}  Cast"
        return f"{show_found}  Cast"

    # otherwise return the cleaned base (pipe suffix removed)
    return base


# Build mapping from raw categories to normalized categories
category_map = {raw: normalize_category(raw) for raw in unique_categories}

# Group raw categories by normalized form for inspection
groups = defaultdict(list)
for raw, norm in category_map.items():
    groups[norm].append(raw)

# Print a short summary
print('Raw categories:', len(unique_categories))
print('Normalized unique categories:', len(groups))
print('\nSample normalized -> count (first 40):')
for i, (norm, raws) in enumerate(list(groups.items())[:40]):
    print(i+1, repr(norm), '->', len(raws))

# Show groups for our three shows (cast groups)
for show in shows.values():
    print(f"\nGroups for {show}:")
    for norm in sorted([k for k in groups.keys() if k.startswith(show)]):
        print(' -', norm, '->', len(groups[norm]))

# Save mapping to data_handling/category_mapping.json
os.makedirs('data_handling', exist_ok=True)
with open(os.path.join('data_handling', 'category_mapping.json'), 'w', encoding='utf-8') as f:
    json.dump(category_map, f, indent=2, ensure_ascii=False)

print('\nSaved mapping at data_handling/category_mapping.json')
```

```python
def clean_categories(raw_categories, category_map):
    """Map raw categories to normalized forms and remove duplicates while preserving order.
    
    Args:
        raw_categories: List of raw category strings
        category_map: Dictionary mapping raw -> normalized category strings
        
    Returns:
        List of unique normalized category strings in order of first appearance
    """
    normalized = []
    seen = set()
    
    for cat in raw_categories:
        # Map to normalized form (use raw if not in mapping)
        norm = category_map.get(cat, cat)
        
        # Only add if we haven't seen this normalized category yet
        if norm not in seen:
            normalized.append(norm)
            seen.add(norm)
    
    return normalized
```

```python
def load_category_mapping(path):
    with open(path, 'r', encoding='utf-8') as f:
        category_map = json.load(f)
    
    return category_map
```
